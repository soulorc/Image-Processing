### Read the image file using scipy.misc
A tutorial for image processing with scipy and numpy can be found [here](http://www.scipy-lectures.org/advanced/image_processing/). 

```python
from scipy import misc

messi = misc.imread('lionel-messi-argentina.jpg')
```

```python
type(messi)
```

```python
messi.shape, messi.dtype
```

#### Display images

```python
import matplotlib.pyplot as plt
plt.imshow(messi)
plt.show()
```

### Import Kmeans class from sklearn

```python
from sklearn.cluster import KMeans
```

### Data Preprocessing
The data is the RGB value of each pixel on the image. The target is to find the cluster centers for all the pixels. We should flatten out the image arrays into (1080*1920,3).

```python
import numpy as np
```

```python
data = np.reshape(messi, (-1,3))
data.shape
```

#### Can we transform it back by simply using np.reshape(data, (1080,1920,3))?

```python
back = np.reshape(data, (1080,1920,3))
```

```python
plt.imshow(back)
plt.show()
```

#### np.reshape(data, (1080,1920,3)) works!

### Let us use the Kmeans algorithm to train the data.

```python
kmeans= KMeans(n_clusters=5)
kmeans.fit(data)
centers = kmeans.cluster_centers_
```

```python
centers
```

```python
predicted = kmeans.predict(data)
```

```python
predicted.shape[0]
```

### The predicted values is the index of the cluster centers. We have to assign the cluster centers to the array.

```python
temp=np.zeros((predicted.shape[0],3))
for i in range(centers.shape[0]):
    temp[predicted==i,:]=centers[i]
```

```python
temp.shape
```

```python
processed=temp.reshape(1080,1920,3)
```

```python
plt.figure(figsize = (20,20))
plt.subplot(1,2,1)
plt.imshow(processed)
plt.subplot(1,2,2)
plt.imshow(messi)
plt.show()
```

## This algorithm essentially compressed the image data size from $ 24N $ to $ 24K + Nlog_2K $ where $N$ is the number of pixels and $K$ is the number of clusters.
